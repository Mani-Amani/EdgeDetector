```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
image = cv2.imread('ezgif-3-0b8625e958.jpg')
```

```python
plt.imshow(image)
plt.show()
```

```python
def image_processing(img, Gauss_flag=1, Color_flag=1, Gray_flag=0): # Image preprocessing 
# Gaussian filter smoothes the image 
    if Gauss_flag == 1:
        img = cv2.GaussianBlur(img, (3, 3), 0)
# Equalize the histogram of color image 
    if Color_flag == 1:
        img = cv2.cvtColor(img, cv2.COLOR_BGR2YUV)
# Convert a color image to a grayscale image , Histogram of balanced gray image 
    if Gray_flag == 1:
        img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        img = cv2.equalizeHist(img)
    return img
    
```

```python
import cv2
def based(): 
    pass
def image_canny(img):
    cv2.namedWindow('Canny')
    cv2.createTrackbar('threshold1', 'Canny', 50, 600, based) # threshold 1
    cv2.createTrackbar('threshold2', 'Canny', 100, 300, based) # Larger threshold 2 It is used to detect obvious edges in the image 
    while(1):
        threshold1 = cv2.getTrackbarPos('threshold1', 'Canny') # threshold 1
        threshold2 = cv2.getTrackbarPos('threshold2', 'Canny') # threshold 2
        img_edges = cv2.Canny(img, threshold1, threshold2)
        cv2.imshow('Canny', img_edges)
        if cv2.waitKey(1) == ord('q'): # Press q sign out 
            break
    cv2.destroyAllWindows()
```

```python
img= image_processing(image)
# Show the original image 
cv2.imshow('Original', image)
# Display the preprocessed image 
cv2.imshow('Pretreatment', img)
image_canny(img)
```
